## Data Preparation

#### Rename Columns

```python
new_names_dict = {'Life Ladder': 'Happiness', 
            'Log GDP per capita': 'LogGDP', 
            'Social support': 'Support', 
            'Healthy life expectancy at birth': 'Life', 
            'Freedom to make life choices': 'Freedom', 
            'Perceptions of corruption': 'Corruption', 
            'Positive affect': 'Positive', 
            'Negative affect': 'Negative',
            'Confidence in national government': 'Confidence',
            'Standard deviation of ladder by country-year': 'STD by year',
            'Standard deviation/Mean of ladder by country-year': 'STD/Mean by year', 
            'GINI index (World Bank estimate)': 'WB estimate',
            'GINI index (World Bank estimate), average 2000-15': 'WB estimate avg',
            'gini of household income reported in Gallup, by wp5-year': 'GINI household'}

df.rename(columns=new_names_dict, inplace=True)
print(df.columns)
```

#### Replace Missing Values

```python
print(df.isnull().sum())
```

```python
# Inspecting numerical and non numerical features

categorical_features = df.select_dtypes(include=['object']).columns.tolist()
numerical_features = df.dtypes.apply(lambda x: x == np.int64 or x == np.float64)

print(categorical_features)
print(numerical_features)
```

```python
# Identifying which columns have missing values 
nan_count = np.sum(df.isnull(), axis=0)
nan_detected = nan_count > 0

print(nan_detected)
```

```python
to_impute = nan_detected & numerical_features
print(to_impute)

# Performing mean imputation on numerical data that is missing: 

## mean_logGDP=df['LogGDP'].mean()
df['LogGDP'].fillna(value=df['LogGDP'].mean(), inplace=True)

## mean_support=df['Support'].mean()
df['Support'].fillna(value=df['Support'].mean(), inplace=True)

## mean_life=df['Life'].mean()
df['Life'].fillna(value=df['Life'].mean(), inplace=True)

## mean_freedom=df['Freedom'].mean()
df['Freedom'].fillna(value=df['Freedom'].mean(), inplace=True)

## mean_corruption=df['Corruption'].mean()
df['Corruption'].fillna(value=df['Corruption'].mean(), inplace=True)

## mean_generosity=df['Generosity'].mean()
df['Generosity'].fillna(value=df['Generosity'].mean(), inplace=True)

## mean_positive=df['Positive'].mean()
df['Positive'].fillna(value=df['Positive'].mean(), inplace=True)

## mean_negative=df['Negative'].mean()
df['Negative'].fillna(value=df['Negative'].mean(), inplace=True)

## mean_confidence=df['Confidence'].mean()
df['Confidence'].fillna(value=df['Confidence'].mean(), inplace=True)

## mean_demqual=df['Democratic Quality'].mean()
df['Democratic Quality'].fillna(value=df['Democratic Quality'].mean(), inplace=True)

## mean_delqual=df['Delivery Quality'].mean()
df['Delivery Quality'].fillna(value=df['Delivery Quality'].mean(), inplace=True)

## mean_wb=df['WB estimate'].mean()
df['WB estimate'].fillna(value=df['WB estimate'].mean(), inplace=True)

## mean_wbavg=df['WB estimate avg'].mean()
df['WB estimate avg'].fillna(value=df['WB estimate avg'].mean(), inplace=True)

## mean_household=df['GINI household'].mean()
df['GINI household'].fillna(value=df['GINI household'].mean(), inplace=True)
```

```python
# Checking imputation results 

for colname in ['LogGDP', 'Support', 'Life', 'Freedom', 'Corruption', 'Generosity', 'Positive', 
                'Negative', 'Confidence', 'Democratic Quality', 'Delivery Quality', 'WB estimate',
                'WB estimate avg', 'GINI household']:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))
```

#### Feature Engineering: One Hot Encoding on Country Column

```python
enc_country = pd.get_dummies(df['country'], prefix='country')
enc_country
```

```python
df = pd.concat([df, enc_country], axis=1)
df.drop('country', axis=1, inplace=True)
```

#### Feature and Label Selection

```python
features = ['Corruption', 'Freedom', 'Generosity', 'Life', 'LogGDP', 'STD by year', 'Support']

X = df[features]
y = df['Confidence_win']
```

#### Create Training and Test Data Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 
```

## Model Selection

#### Cross Validation Across Candidates 

```python
estimators = [("DT", DecisionTreeRegressor(max_depth=8, min_samples_leaf=25)),
              ("LR", LinearRegression())
             ]

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(max_depth=32, n_estimators=300),
    "Gradient Boosting": GradientBoostingRegressor(max_depth=2, n_estimators=300),
    "Stacking Regressor": StackingRegressor(estimators=estimators, passthrough=False)
}

RMSE_results = []
R2_results = []

for name, model in models.items():
    mse_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    RMSE_results.append(np.mean(-mse_scores)) 

    r2_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    R2_results.append(np.mean(r2_scores))

```

#### Visualize Results

```python
rg= np.arange(5)
width = 0.35

# 1. Create bar plot with RMSE results
plt.bar(rg, RMSE_results, width, label='RMSE')


# 2. Create bar plot with R2 results
plt.bar(rg + width, R2_results, width, label='R2')



labels = ['LR', 'DT', 'RF', 'GDBT', 'Stacking']
plt.xticks(rg + width/2, labels)

plt.xlabel("Models")
plt.ylabel("RMSE/R2")


plt.ylim([0,1])
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()
```

## Random Forest Model Optimization and Evaluation

```python
# Building and cross validating model 

rf_model = RandomForestRegressor(random_state=42)

rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Test MSE: {mse}")
print(f"Test RÂ²: {r2}")

cv_mse_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
cv_r2_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='r2')
print(f"Cross-validated MSE: {np.mean(-cv_mse_scores)}")
print(f"Cross-validated RÂ²: {np.mean(cv_r2_scores)}")
```

```python
# Optimizing hyperparameters

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
best_rf_model = grid_search.best_estimator_
```

### Evaluation Using Visualizations

```python
# Testing model with new parameters 

y_pred_tuned = best_rf_model.predict(X_test)
mse_tuned = mean_squared_error(y_test, y_pred_tuned)
r2_tuned = r2_score(y_test, y_pred_tuned)

print(f"Tuned Test MSE: {mse_tuned}")
print(f"Tuned Test RÂ²: {r2_tuned}")
```
